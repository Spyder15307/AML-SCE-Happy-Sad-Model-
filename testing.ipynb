```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
test_model = load_model(os.path.join("models", "happy_sad_model.h5"))
```

```python
test_img = cv2.imread(os.path.join("uploads", "sadtest3.jpg"))
plt.imshow(test_img)
plt.show()
```

```python
test_resize = tf.image.resize(test_img, (256,256))
```

```python
prediction = test_model.predict(np.expand_dims(test_resize/255, 0))
```

```python
if prediction < 0.5:
    print("Person is Happy")
else:
    print("Person is sad")
```
